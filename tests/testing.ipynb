```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample: House sizes (x), prices (y) with noise
X = np.array([[1], [2], [3], [4], [5]])  # sizes (add bias: np.c_[np.ones(5), X])
y = np.array([2.1, 4.0, 6.2, 7.9, 10.1])  # prices (true: 2*size, + Gaussian noise)

model = LinearRegression(fit_intercept=True)  # Handles bias
model.fit(X, y)
theta = np.hstack([model.intercept_, model.coef_])  # [bias, slope]
print(f"Learned θ: {theta}")  # e.g., [0.1, 1.98] ≈ true [0, 2]

# Predict & plot
plt.scatter(X, y, color='blue', label='Data (noisy)')
plt.plot(X, model.predict(X), color='red', label='Fit line')
plt.xlabel('House Size'); plt.ylabel('Price'); plt.legend(); plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Dataset: hours_studied (feature), passed (0/1)
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([0, 0, 1, 1, 1])

model = LogisticRegression()  # Default SGD-like solver
model.fit(X, y)
probs = model.predict_proba(X)[:, 1]  # P(y=1)
preds = model.predict(X)  # 0/1 decisions
print(f"Probs: {probs}")  # e.g., [0.3, 0.45, 0.65, 0.8, 0.9]
print(f"Accuracy: {accuracy_score(y, preds)}")  # 1.0

# New prediction: 2.5 hours?
print(f"P(pass|2.5 hrs): {model.predict_proba([[2.5]])[:,1][0]}")
```
